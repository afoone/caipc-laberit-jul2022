# univariate

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import math

```

```python
def knn_train_test(train_col, target_col, df):
    """
    Calcule la predicción y me devuelva el RMSE en base a una serie de columnas 
    predictoras y una target a predecir
    """
    # Crearemos la instancia del objeto de KNeighborsRegressor
    knn = KNeighborsRegressor()
    
    # Barajaremos las filas en el dataframe
    np.random.seed(1)
    coches_random = coches_numericos.reindex(np.random.permutation(coches_numericos.index))
    # Crearemos los conjuntos de entrenamiento y de test
    split_value = math.trunc(len(coches_random)*0.8)
    train_df = coches_random.iloc[0:split_value]
    test_df = coches_random.iloc[split_value:]
    
    # Haremos el fit del model usando k=5
    knn.fit(train_df[[train_col]], train_df[target_col])
    
    # Crearemos las predicciones 
    predictions = knn.predict(test_df[[train_col]])
    # Calcularemos y devolveremos el rmse
    mse = mean_squared_error(test_df[[target_col]], predictions)
    
    return mse**0.5
    
```

```python
# Calcular el RMSE de cada una de las columnas numéricas
# veremos qué columnas nos predice mejor el precio
for col in ['symboling', 'normalized-losses', 'num-of-doors', 'wheel-base', 'engine-size', 'bore',
                          'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']:
    print('col', col)
    print(col, knn_train_test(col, 'price', coches_numericos))
```

```python
# multivariate
# comprobaremos varias combinaciones, incluyendo la de 'todas las columnas'
def knn_train_test_multi(train_cols, target_col, df):
    """
    Calcule la predicción y me devuelva el RMSE en base a una serie de columnas 
    predictoras y una target a predecir
    """
    # Crearemos la instancia del objeto de KNeighborsRegressor
    knn = KNeighborsRegressor()
    
    # Barajaremos las filas en el dataframe
    np.random.seed(1)
    coches_random = coches_numericos.reindex(np.random.permutation(coches_numericos.index))
    # Crearemos los conjuntos de entrenamiento y de test
    split_value = math.trunc(len(coches_random)*0.8)
    train_df = coches_random.iloc[0:split_value]
    test_df = coches_random.iloc[split_value:]
    
    # Haremos el fit del model usando k=5
    knn.fit(train_df[train_cols], train_df[target_col])
    
    # Crearemos las predicciones 
    predictions = knn.predict(test_df[train_cols])
    # Calcularemos y devolveremos el rmse
    mse = mean_squared_error(test_df[[target_col]], predictions)
    
    return mse**0.5
```

```python
knn_train_test_multi(['engine-size', 'highway-mpg', 'horsepower'], 'price', coches_numericos)
```

```python
# multivariate
# comprobaremos varias combinaciones, incluyendo la de 'todas las columnas'
def knn_final(train_cols, target_col, df):
    """
    Calcule la predicción y me devuelva el RMSE en base a una serie de columnas 
    predictoras y una target a predecir
    """

    
    # Barajaremos las filas en el dataframe
    np.random.seed(1)
    coches_random = coches_numericos.reindex(np.random.permutation(coches_numericos.index))
    # Crearemos los conjuntos de entrenamiento y de test
    split_value = math.trunc(len(coches_random)*0.8)
    train_df = coches_random.iloc[0:split_value]
    test_df = coches_random.iloc[split_value:]
    
    k_values = range(1,10)
    k_rmse = {}
    
    for k in k_values:
        # Crearemos la instancia del objeto de KNeighborsRegressor
        knn = KNeighborsRegressor(n_neighbors = k)
        # Haremos el fit del model usando k=k
        knn.fit(train_df[train_cols], train_df[target_col])
        # Crearemos las predicciones 
        predictions = knn.predict(test_df[train_cols])
        # Calcularemos y devolveremos el rmse
        mse = mean_squared_error(test_df[[target_col]], predictions)
        k_rmse[k] = mse**0.5
    
    return k_rmse
```

```python
# probaremos varios valores de k
knn_final(['engine-size',  'horsepower'], 'price', coches_numericos)
```
