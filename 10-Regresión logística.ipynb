```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/afoone/caipc-laberit-jul2022/master/titanic.csv')
```

```python
 titanic = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
titanic
```

```python
titanic.info()
```

```python
gender_dummies = pd.get_dummies(titanic['Sex'], prefix='gender')
```

```python
embarked_dummies = pd.get_dummies(titanic['Embarked'], prefix='embarked')
# categorizar la variable p_class
class_dummies = pd.get_dummies(titanic['Pclass'], prefix='class')
titanic = titanic.join(gender_dummies).join(embarked_dummies).join(class_dummies)
```

```python
titanic
```

```python
titanic = titanic.drop(['Sex', 'Embarked', 'Pclass'], axis=1)
```

```python
titanic
```

# Seleccionar variables

```python
from sklearn import datasets
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
# instanciamos la regresión logística
lr = LogisticRegression()
```

```python
titanic.columns
```

```python
titanic_dropna = titanic.dropna()
```

```python
titanic_dropna.info()
```

```python
# hacer una feature selection
rfe = RFE(lr, n_features_to_select=4)
rfe.fit(titanic_dropna[['Age', 'SibSp', 'Parch', 'Fare', 'gender_female',
       'gender_male', 'embarked_C', 'embarked_Q', 'embarked_S', 'class_1',
       'class_2', 'class_3']], titanic_dropna['Survived'])
```

```python
titanic_dropna['class_3'].sum()
```

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
z = zip(['Age', 'SibSp', 'Parch', 'Fare', 'gender_female',
       'gender_male', 'embarked_C', 'embarked_Q', 'embarked_S', 'class_1',
       'class_2', 'class_3'], rfe.support_, rfe.ranking_)
list(z)
```

```python
lr.fit(titanic_dropna[['gender_male', 'gender_female', 'class_1', 'class_3']], titanic_dropna['Survived'])
```

```python
lr.score(titanic_dropna[['gender_male', 'gender_female', 'class_1', 'class_3']], titanic_dropna['Survived'])
```

```python
titanic_woage = titanic.dropna(axis = 1)
```

```python
titanic_woage
```

```python
# hacer una feature selection
rfe = RFE(lr, n_features_to_select=4)
rfe.fit(titanic_woage[['SibSp', 'Parch', 'Fare', 'gender_female',
       'gender_male', 'embarked_C', 'embarked_Q', 'embarked_S', 'class_1',
       'class_2', 'class_3']], titanic_woage['Survived'])
```

```python
z = zip(['SibSp', 'Parch', 'Fare', 'gender_female',
       'gender_male', 'embarked_C', 'embarked_Q', 'embarked_S', 'class_1',
       'class_2', 'class_3'], rfe.support_, rfe.ranking_)
list(z)
```

```python
lr_full = LogisticRegression()
lr_full.fit(titanic_woage[[ 'gender_female',
       'gender_male','class_1',
       'class_3']], titanic_woage['Survived'])
lr_full.score(titanic_woage[[ 'gender_female',
       'gender_male','class_1',
       'class_3']], titanic_woage['Survived'])
```

# validación

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(titanic_woage, test_size=0.2, random_state=0)
```

```python
train.shape
```

```python
test.shape
```

```python
lr_full = LogisticRegression()
lr_full.fit(train[[ 'gender_female',
       'gender_male','class_1',
       'class_3']], train['Survived'])
lr_full.score(train[[ 'gender_female',
       'gender_male','class_1',
       'class_3']], train['Survived'])
```

```python
test['predicted'] = lr_full.predict(test[[ 'gender_female',
       'gender_male','class_1',
       'class_3']])
```

```python
test[test['Survived']==test['predicted']]['predicted'].count() / len(test)
```

```python
lr_full.predict(test[[ 'gender_female',
       'gender_male','class_1',
       'class_3']])
```

```python
lr_full.predict_proba(
    test[[ 'gender_female',
       'gender_male','class_1',
       'class_3']]
)[:,1]
```

```python
test['pro_surv'] = lr_full.predict_proba(
    test[[ 'gender_female',
       'gender_male','class_1',
       'class_3']]
)[:,1]
```

```python
test[test['pro_surv']>0.89]
```
