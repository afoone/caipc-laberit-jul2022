```python
def dibujar_secante(x_values):
    x = np.linspace(-20, 30, 100)
    y = -1*x**2 + x*3 -1
    plt.plot(x, y)
    x_0 = x_values[0]
    x_1 = x_values[1]
    y_0 = -1*x_0**2 + x_0*3 -1
    y_1 = -1*x_1**2 + x_1*3 -1
    
    m = (y_1 - y_0)/(x_1 - x_0)
    
    b = y_1 - m*x_1
    
    
    y_secante = m*x + b
    
    plt.plot(x, y_secante, color='green')
    
    plt.show()
    
```

```python
dibujar_secante([5, 5])
```

```python
dibujar_secante([-10, 20])
```

```python
dibujar_secante([0, 10])
```

```python
dibujar_secante([4.999,5.0001 ])
```
